## Step 2: Import data

This notebook assumes you are working with the dataframe created in the last step. Don't forget to change the path to match your working directory!

```python
df = pd.read_csv('/content/drive/My Drive/THISTLE_project/post_ocr_output.csv')
```

## Step 3: Create a list of columns to compare

```python
ocr_cols = ['ocr', 'tesseract_ocr', 'llama_output', 'llama_fine_tuned_output'] # change this list as needed to match your datset
```

## Step 4: Calculate WER and CER

This code block calculates WER and CER for each row in each comparison column, comparing it to the ground truth in the 'manual_transcription' column.

```python

# Compute WER and CER for each comparison column
for col in ocr_cols:
    df[f'{col}_wer'] = df.apply(lambda row: wer(row['manual_transcription'], row[col]), axis=1)
    df[f'{col}_cer'] = df.apply(lambda row: cer(row['manual_transcription'], row[col]), axis=1)

print(df.head())
```
